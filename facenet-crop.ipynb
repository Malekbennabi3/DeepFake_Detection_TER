```python
import os
import shutil

# copier toutes les images d’un dossier source vers un dossier de destination. 
def copy_images(source_folder, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Walk through each subfolder in the source folder
    for root, dirs, files in os.walk(source_folder):
        for file in files:
            # Check if the file is an image
            if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
                # Get the full path of the image
                source_path = os.path.join(root, file)
                # Copy the image to the destination folder
                shutil.copy(source_path, destination_folder)
```

```python
# Specify the source folder containing subfolders with images
destination_folder = 'real'
# Specify the destination folder where images will be copied
source_folder = 'wiki'

# Call the function to copy images
copy_images(source_folder, destination_folder)
```

```python
# Specify the source folder containing subfolders with images
destination_folder = 'fake'
# Specify the destination folder where images will be copied
source_folder = 'inpainting'

# Call the function to copy images
copy_images(source_folder, destination_folder)
```

```python
import os
from PIL import Image
from mtcnn import MTCNN
import shutil
import numpy as np
import matplotlib.pyplot as plt
```

```python
#!zip -r "C:\Users\HP\Downloads\face_fake.zip" "face_fake"
```

```python
# Create the detector, using default weights
detector = MTCNN()

# Chemin vers le dossier contenant les images
folder_path = '/kaggle/working/fake'

# Chemin vers le dossier de sortie
output_folder = '/kaggle/working/face_fake'


# Créer le dossier de sortie s'il n'existe pas
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Parcourir tous les fichiers du dossier
for filename in os.listdir(folder_path):
    # Vérifier si le fichier est une image
    if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
        # Ouvrir l'image
        image = Image.open(os.path.join(folder_path, filename))
        
        # Redimensionner l'image à 224x224 pixels
        image = image.resize((224, 224))
        
        # Convertir l'image en mode RGB
        image = image.convert('RGB')
        
        # Convertir l'image en un tableau numpy
        pixels = np.asarray(image)
        
        # Detect faces in the image
        results = detector.detect_faces(pixels)

        if results:
            # Extract the bounding box from the first face
            x, y, width, height = results[0]['box']
            # Adjust the bounding box
            x, y = abs(x), abs(y)
            face = pixels[y:y+height, x:x+width]
            
            # Parcourir tous les visages détectés
            for i, result in enumerate(results):
            # Obtenir les coordonnées du visage
                x1, y1, width, height = result['box']
                x2, y2 = x1 + width, y1 + height

            # Extraire le visage
                face = pixels[y1:y2, x1:x2]

            # Créer une image à partir du visage
                image_face = Image.fromarray(face)

            # Enregistrer le visage dans le dossier de sortie avec un nom unique
                output_file = os.path.join(output_folder, f"{filename}_face{i}.png")
                image_face.save(output_file)
```
